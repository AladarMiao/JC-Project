```python
import pandas as pd
from IPython.display import display
from sklearn.impute import SimpleImputer

class DataPreprocessor:
    def __init__(self, filename):
        self.df = pd.read_csv(filename)
        display(self.df)
        self.X = self.df.iloc[:, :-1].values
        self.y = self.df.iloc[:, -1].values
        self.remove_duplicates_option()
        self.impute_values_option()
    
    def remove_duplicates_option(self):
        remove_duplicates = input("Do you want to remove duplicates? Input T or F ") == "T"
        if remove_duplicates:
            self.df.drop_duplicates(inplace=True)
            
    def impute_values_option(self):
        impute_values = input("Do you want to impute values? Input T or F ") == "T"
        if impute_values:
            impute_strategy = input("How do you want to impute your values? e.g. \"median\", 0 " )
            imputer = SimpleImputer()
            if impute_strategy.isdigit():
                imputer = SimpleImputer(strategy='constant', fill_value=int(impute_strategy))
            else:
                imputer = SimpleImputer(strategy = impute_strategy)
            self.df = pd.DataFrame(imputer.fit_transform(self.df), columns=self.df.columns)
    
    def display_table(self):
        display(self.df)
```

```python
preprocessor = DataPreprocessor(filename='Median Earnings for New CS Grads.csv')
```

```python
preprocessor.display_table()
```

```python
def binary_to_decimal(s):
    ret_value = 0
    for index in range(len(s)-1, -1, -1):
        if s[index]=="1":
            ret_value+=2**(len(s)-1-index)

    return ret_value
```

```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score

# Load the CSV file
data = pd.read_csv('Median Earnings for New CS Grads.csv')

# Define a function to preprocess the data
def preprocess_data(data, remove_duplicates=False, fill_nan_with=0):
    if remove_duplicates:
        data.drop_duplicates(inplace=True)
    if fill_nan_with is not None:
        data.fillna(fill_nan_with, inplace=True)
    return data

# Define a function to perform clustering
def perform_clustering(data, algorithm='kmeans', n_clusters=3):
    if algorithm == 'kmeans':
        model = KMeans(n_clusters=n_clusters)
    elif algorithm == 'dbscan':
        model = DBSCAN()
    else:
        raise ValueError('Invalid clustering algorithm')
    
    clusters = model.fit_predict(data)
    return clusters

# Define a function to measure performance
def measure_performance(data, clusters):
    score = silhouette_score(data, clusters)
    return score

# Define a loop to allow the user to interact with the program
while True:
    # Ask the user for preprocessing options
    remove_duplicates = input('Remove duplicates? (y/n) ') == 'y'
    fill_nan_with = input('Fill NaN values with what? (enter a number or leave blank for none) ')
    if fill_nan_with != '':
        fill_nan_with = float(fill_nan_with)
    
    # Preprocess the data
    data = preprocess_data(data, remove_duplicates, fill_nan_with)
    
    # Ask the user for clustering options
    algorithm = input('Which clustering algorithm? (kmeans/dbscan) ')
    n_clusters = int(input('How many clusters? '))
    
    # Perform clustering
    clusters = perform_clustering(data, algorithm, n_clusters)
    
    # Measure performance
    score = measure_performance(data, clusters)
    
    # Print the score
    print('Silhouette score: ', score)
    
    # Ask the user if they want to continue
    if input('Do you want to continue? (y/n) ') != 'y':
        break
```
